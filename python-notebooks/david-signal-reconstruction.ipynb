```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# load files 3.3V CSV
broad_signal_3_3 = pd.read_csv('../data/broad.Wfm.csv', delimiter=';', names=['near_end', 'far_end'])
strip_signal_3_3 = pd.read_csv('../data/strip.Wfm.csv', delimiter=';', names=['near_end', 'far_end'])
ustrip_signal_3_3 = pd.read_csv('../data/broad.Wfm.csv', delimiter=';', names=['near_end', 'far_end'])
```

```python
################## 3.3V plots ##################

time_start = 32
window_length = 125000
time_delta = (0.000140214-0.000140014)*1e9

###### Microstrup/ustrip ######
ustrip_start_idx = int(time_start * 500000/200)
ustrip_end_idx = ustrip_start_idx + window_length
x_ustrip = np.linspace(0, time_delta * window_length/500000, window_length)

###### Stripline ######
strip_start_idx = int(time_start * 500000/200)
strip_end_idx = ustrip_start_idx + window_length
x_strip = np.linspace(0, time_delta * window_length/500000, window_length)

###### Broadside #####
broad_start_idx = int(time_start * 500000/200)
broad_end_idx = ustrip_start_idx + window_length
x_broad = np.linspace(0, time_delta * window_length/500000, window_length)

"""
broad_signal_3_3_start = 0
broad_signal_3_3_end = (0.000140214-0.000140014)*1e9
broad_signal_3_3_length = broad_signal_3_3.shape[0]

broad_signal_3_3_linspace = np.linspace(broad_signal_3_3_start, broad_signal_3_3_end,
                                        num=broad_signal_3_3_length)
"""

##### Plots #####
fig, ax = plt.subplots(2,3)
fig.set_figheight(7)
fig.set_figwidth(12)

##### Common stuff #####
ax[0,0].set_ylabel('Near End')
ax[1,0].set_ylabel('Far End')

###### Microstrup/ustrip ######
ax[0,0].plot(x_ustrip, ustrip_signal_3_3['near_end'][ustrip_start_idx:ustrip_start_idx+window_length]*1000, color='blue', linewidth=0.75)
ax[1,0].plot(x_ustrip, ustrip_signal_3_3['far_end'][ustrip_start_idx:ustrip_start_idx+window_length]*1000, color='orange', linewidth=0.75)

ax[0,0].grid()
ax[1,0].grid()
ax[0,0].set_title('Microstrip')
ax[0,0].set_ylim([-200, 235])
ax[1,0].set_ylim([-200, 235])

###### Microstrup/ustrip ######
ax[0,1].plot(x_strip, strip_signal_3_3['near_end'][strip_start_idx:strip_start_idx+window_length]*1000, color='blue', linewidth=0.75)
ax[1,1].plot(x_strip, strip_signal_3_3['far_end'][strip_start_idx:strip_start_idx+window_length]*1000, color='orange', linewidth=0.75)

ax[0,1].grid()
ax[1,1].grid()
ax[0,1].set_title('Stripline')
ax[0,1].set_ylim([-200, 235])
ax[1,1].set_ylim([-200, 235])

###### Broadside ######
ax[0,2].plot(x_broad, broad_signal_3_3['near_end'][broad_start_idx:broad_start_idx+window_length]*1000, color='blue', linewidth=0.75)
ax[1,2].plot(x_broad, broad_signal_3_3['far_end'][broad_start_idx:broad_start_idx+window_length]*1000, color='orange', linewidth=0.75)

ax[0,2].grid()
ax[1,2].grid()
ax[0,2].set_title('Broadside')
ax[0,2].set_ylim([-200, 235])
ax[1,2].set_ylim([-200, 235])


##### Common #####
fig.supylabel('Voltage [mV]')
fig.supxlabel('Time [ns]')
plt.show()
```
